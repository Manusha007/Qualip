```python
# parse BRG to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#show image
plt.imshow(image)
```

```python
WIDTH = 300
HEIGHT = 300
```

```python
#resize, ignoring aspect ration
resized = cv2.resize(image, (WIDTH, HEIGHT))

#show image
plt.imshow(resized)
```

```python
aspect = image.shape[1]/float(image.shape[0])
print(aspect)

if(aspect >1):
    #landscape orientation - wide image
    res = int(aspect * HEIGHT)
    scaled = cv2.resize(image, (res, HEIGHT))
if (aspect <1):
    #portrait orientation - tall image
    res = int(WIDTH/ aspect)
    scaled = cv2.resize(image, (WIDTH, res))
if (aspect ==1):
     scaled = cv2.resize(image, (WIDTH, HEIGHT))
        
#show image
plt.imshow(scaled)
    
```

```python
def crop_center(img, cropx, cropy):
    y,x,c = img.shape
    startx = x//2-(cropx//2)
    starty = y//2-(cropy//2)
    return img[starty:starty+cropy, startx:startx+cropy]

#scaled image
cropped = crop_center(scaled, WIDTH, WIDTH)

#show image
plt.imshow(cropped, cmap='gray')
```

```python
image = cropped.copy()

#add text
cv2.putText(image, "No damage", (70,270),
            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,0,0),2)
#show image
plt.imshow(image, cmap='gray')
```
